```python
import torch
from model import ASRModel2

model = ASRModel2()

# model overview
# 20210715_12_58_50_PM_NAQVGNY trained with airbus dataset
# 20210714_12_36_37_PM_JVMDKLO trained with librispeech
# 20210715_02_28_06_PM_AGFWKNL trained with librispeech, finetuned with airbus
model_name_dict = {
    1: "20210715_12_58_50_PM_NAQVGNY", # trained with airbus dataset
    2: "20210714_12_36_37_PM_JVMDKLO", # trained with librispeech
    3: "20210715_02_28_06_PM_AGFWKNL", # trained with librispeech, finetuned with airbus
    4: "20210715_07_40_43_PM_CQJNDXX" # trained with mixed librispeech and airbus
}

model.load_state_dict(torch.load(
    "saved_models/"+model_name_dict[4])["model_state_dict"]
)
```

```python
from data import play_waveform, target_to_text, prepare_waveform, play_waveform
from airbus_dataset import AirbusDataset
from model import predict

dataset = AirbusDataset()
waveform, sample_frequency, utterance, _, _, _ = dataset.__getitem__(7)

model.eval()
input, input_length = prepare_waveform(waveform, sample_frequency, 16000)
output = model(input.to(0))
pred = predict(output, input_length)
print(utterance)
print()
print(target_to_text(*pred))


play_waveform(waveform, sample_frequency)
```
